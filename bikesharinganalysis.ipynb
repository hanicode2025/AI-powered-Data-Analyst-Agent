```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/bike_sharing.csv')
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['datetime'] = pd.to_datetime(data['datetime'])
```

```python
data['hour'] = data['datetime'].dt.hour
data['hour'].head(24)
```

```python
hr_demand = data.groupby('hour')['count'].mean()
```

```python
sns.barplot(x=hr_demand.index, y=hr_demand.values)
plt.title('Average Bike Demand per Hour')
plt.xlabel('Hour')
plt.ylabel('Average Count')
plt.show()
```

```python
data['season'].value_counts()
```

```python
sns.barplot(x='season',y='count',data=data)
plt.title('Season Count')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x = 'temp', y = 'count', data = data, hue = 'season')
plt.title('Temperature vs Count')
plt.xlabel('Temperature')
plt.ylabel('Count')
plt.show()
```

```python
sns.barplot(x = 'weather', y = 'count', data = data, hue = 'season')
plt.title('Weather vs Count')
plt.xlabel('Weather')
plt.ylabel('Count')
plt.show()
```
